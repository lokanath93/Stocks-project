```python
import requests
from bs4 import BeautifulSoup
import csv

def fetch_and_write_to_csv(url, table_class, csv_filename,sector,set_header):
    # Make a request to the webpage
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code != 200:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return
    
    # Parse the HTML content of the webpage
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find the table with the specific class name
    table = soup.find('table', {'class': table_class})
    
    if table is None:
        print("Table with the specified class name not found.")
        return
    
    # Extract table headers
    headers = [header.text.strip() for header in table.find_all('th')]
    headers.append('sectors')
    #print(headers)
    # Extract table rows
    rows = table.find_all('tr')
    data = []
    for row in rows:
        columns = row.find_all('td')
        #print(columns)
        if columns:
            #data.append(sector)
            lst_rows = [column.text.strip() for column in columns]
            lst_rows.append(sector)
            data.append(lst_rows)
        
    # Write data to a CSV file
    with open(csv_filename, mode='a', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        # Write the headers
        if set_header:
            writer.writerow(headers)
        # Write the data
        writer.writerows(data)
    
    print(f"Data written to {csv_filename} for sector {sector}")

# Example usage
url = 'https://moneycontrol.com/india/stockmarket/sector-classification/marketstatistics/nse/agri.html'
table_class = 'tbldata14 bdrtpg'  # Replace with the actual class name of the table
csv_filename = 'scraped_data.csv'

#fetch_and_write_to_csv(url, table_class, csv_filename, 'qwer', True)
```

```python
import requests
from bs4 import BeautifulSoup
import time

def get_li_under_div(url, div_class, date):
    # Make a request to the webpage
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code != 200:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        return []
    
    # Parse the HTML content of the webpage
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find the div with the specific class name
    div = soup.find('div', {'class': div_class})
    
    if div is None:
        print("Div with the specified class name not found.")
        return []
    
    # Extract all li elements under the div
    li_elements = div.find_all('li')
    
    # Extract text from each li element
    #li_texts = [li.text.strip() for li in li_elements]
    base_url = "https://moneycontrol.com"
    set_header = True
    for li in li_elements:
        a_element = li.find('a')
        url = base_url
        if a_element and 'href' in a_element.attrs:
            url = url + a_element['href']
        if url != base_url:
            sector = li.text.strip()
            #print(sector)
            fetch_and_write_to_csv(url, table_class, date+'_'+csv_filename, sector, set_header)
            set_header = False
        li_texts.append(li.text.strip)
    
    return li_texts

# Example usage
url = 'https://moneycontrol.com/india/stockmarket/sector-classification/marketstatistics/nse/agri.html'
div_class = 'lftmenu'  # Replace with the actual class name of the div
li_texts = get_li_under_div(url, div_class, time.strftime("%Y%m%d"))
#for text in li_texts:
#    print(text)
```
